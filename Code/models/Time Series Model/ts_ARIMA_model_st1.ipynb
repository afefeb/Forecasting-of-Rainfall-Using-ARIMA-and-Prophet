```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.dates as mdates
```

Time Series Forecasting for Average Temperature

```python
df_stasiun1 = pd.read_csv('C:\\Users\\fayme amelia celesta\\Documents\\VScode\\Github\\Intern-BCC-2024\Dataset\\Processed\\time series\\stasiun_1_intrp.csv', index_col="Unnamed: 0")
```

```python
df_stasiun1
```

```python
df_stasiun1.index = pd.to_datetime(df_stasiun1.index)
```

```python
def train(column_name, p, d, q) :
    title = column_name + ' Forecast'
    train_data = df_stasiun1[column_name][:len(df_stasiun1)-365]
    test_data = df_stasiun1[column_name][len(df_stasiun1)-365:]
    
    model = ARIMA(train_data, order=(p, d, q))  
    model = model.fit()

    forecast = model.forecast(steps=365)

    # Hitung mean squared
    mae = mean_absolute_error(test_data, forecast)
    mse = mean_squared_error(test_data, forecast)
    print('Mean Absolute Error:', mae)

    fig = plt.figure(figsize=(10,4))
    plt.plot(train_data.index, train_data, label='Train')
    plt.plot(test_data.index, test_data, label='Test')
    plt.plot(forecast.index, forecast, label='Forecast')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
train('Tavg', 1, 0, 1)
```

```python
train('RH_avg', 1, 0, 1)
```

```python
train('ss', 2, 0, 1)
```

```python
train('RR', 2,0,1)
```

```python
train('ff_avg', 2,1,1)
```

```python
train('ddd_x', 2,1,1)
```

```python
df_stasiun1_copy = df_stasiun1.copy()
df_stasiun1_copy['ddd_x'] = df_stasiun1_copy['ddd_x'].apply(lambda x: x * np.pi / 180)
```

```python
title = 'ddd_x Radiant Forecast'
train_data = df_stasiun1_copy['ddd_x'][:len(df_stasiun1_copy)-365]
test_data = df_stasiun1_copy['ddd_x'][len(df_stasiun1_copy)-365:]
    
model = ARIMA(train_data, order=(2,1,1))  
model = model.fit()

forecast = model.forecast(steps=365)

# Hitung mean squared
mae = mean_absolute_error(test_data, forecast)
mse = mean_squared_error(test_data, forecast)
print('Mean Absolute Error:', mae)

fig = plt.figure(figsize=(10,4))
plt.plot(train_data.index, train_data, label='Train')
plt.plot(test_data.index, test_data, label='Test')
plt.plot(forecast.index, forecast, label='Forecast')
plt.title(title)
plt.legend()
plt.show()
```
